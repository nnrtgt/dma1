```python
#https://stackoverflow.com/questions/29481485/creating-a-distance-matrix
import pandas as pd
from scipy.spatial import distance_matrix
    
data = [[5, 7], [7, 3], [8, 1]]
ctys = ['A', 'B', 'C']
df = pd.DataFrame(data, columns=['xcord', 'ycord'], index=ctys)
```

```python
#Euclidean Distance Matrix
pd.DataFrame(distance_matrix(df.values, df.values), index=df.index, columns=df.index)
```

```python
#Manhattan Distance Matrix
pd.DataFrame(distance_matrix(df.values, df.values,1), index=df.index, columns=df.index)
```

```python
import pandas as pd
from scipy.spatial import distance_matrix

data = [[1, 2], [1, 4], [2, 4], [3,2], [3,5],[3,7],[4,5],[4,6]]
ctys = ['A', 'B', 'C','D','E','F','G','H']
df = pd.DataFrame(data, columns=['xcord', 'ycord'], index=ctys)


#Manhattan Distance Matrix
pd.DataFrame(distance_matrix(df.values, df.values,1), index=df.index, columns=df.index)
```

```python
#Euclidean Distance Matrix
pd.DataFrame(distance_matrix(df.values, df.values), index=df.index, columns=df.index)
```

```python
import math
#Supremum Distance Matrix
pd.DataFrame(distance_matrix(df.values, df.values,math.inf), index=df.index, columns=df.index)
```
